# **STEP 1: IMPORT LIBRARIES**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **STEP 2: LOAD DATA**

```python
df = pd.read_csv("/content/credit_score.csv")
df
```

# **STEP 3: DATA CLEANING**

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Remove unwanted columns if exist
drop_cols = ['ID','Customer_ID','Name','SSN','Type_of_Loan']
```

```python
df.drop(columns=drop_cols, errors='ignore', inplace=True)
```

```python
df['Occupation'].unique()
```

```python
df.head()
```

```python
# Convert Age to numeric
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
```

```python
# Fix wrong age values
df["Age"] = df["Age"].apply(lambda x: np.nan if (pd.isna(x) or x < 0 or x > 100) else x)
```

```python
# Replace wrong values in Credit Mix
df["Credit_Mix"] = df["Credit_Mix"].replace("_", "Unknown")
```

```python
df.info()
```

```python
df.dtypes
```

```python
cols_to_convert = [
    "Num_of_Loan",
    "Num_of_Delayed_Payment",
    "Changed_Credit_Limit",
    "Amount_invested_monthly"
]

for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
for col in cols_to_convert:
    df[col] = df[col].fillna(0).astype(int)
```

```python
df.dtypes
```

# **STEP 4: SPLIT X & y**

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
```

```python
X = df.drop("Credit_Score", axis=1)
y = df["Credit_Score"]
```

```python
X
```

```python
y
```

```python
# Categorical Columns
cat_cols = X.select_dtypes(include=["object"]).columns
num_cols = X.select_dtypes(include=["int64", "float64"]).columns
```

```python
# Convert ALL categorical to string SAFELY
X[cat_cols] = X[cat_cols].astype(str)
```

```python
# Target ko bhi ek hi type banao
y = y.astype(str)
```

# **STEP 5: PREPROCESSING**

```python
numeric_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median"))
])
```

```python
numeric_pipeline
```

```python
categorical_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])
```

```python
categorical_pipeline
```

```python
preprocessor = ColumnTransformer([
    ("num", numeric_pipeline, num_cols),
    ("cat", categorical_pipeline, cat_cols)
])
```

```python
preprocessor
```

# **STEP 6: RANDOM FOREST MODEL**

```python
rf = RandomForestClassifier(random_state=42)
```

```python
pipeline = Pipeline([
    ("preprocess", preprocessor),
    ("rf", rf)
])
```

```python
pipeline
```

# **STEP 7: TRAIN TEST SPLIT**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
X_test
```

```python
X_train
```

```python
y_test
```

```python
y_train
```

```python
categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])
```

# **STEP 8: HYPERPARAMETER TUNING**

```python
param_grid = {
    "rf__n_estimators": [100, 200],
    "rf__max_depth": [8, 10, 12],
    "rf__min_samples_split": [2, 5],
    "rf__min_samples_leaf": [1, 2]
}
```

```python
param_grid
```

```python
grid = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=3,
    scoring="accuracy",
    n_jobs=-1
)
```

```python
grid
```

```python
grid.fit(X_train, y_train)
```

```python

best_model = grid.best_estimator_
print("Best Parameters:", grid.best_params_)
```

# **STEP 9: PREDICTIONS**

```python
y_pred = best_model.predict(X_test)
```

```python
print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

# **STEP 10: FEATURE IMPORTANCE**

```python
# Get feature names after encoding
ohe = best_model.named_steps["preprocess"].named_transformers_["cat"]["encoder"]
```

```python
ohe
```

```python
encoded_cat_features = ohe.get_feature_names_out(cat_cols)
```

```python
encoded_cat_features
```

```python
final_feature_names = np.concatenate([num_cols, encoded_cat_features])
```

```python
final_feature_names
```

```python
importances = best_model.named_steps["rf"].feature_importances_
```

```python
importances
```

```python
feature_importance_df = pd.DataFrame({
    "Feature": final_feature_names,
    "Importance": importances
}).sort_values(by="Importance", ascending=False)
```

```python
feature_importance_df
```
